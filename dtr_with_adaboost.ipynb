```python
import numpy as np

rng = np.random.RandomState(1)
X = np.linspace(0, 6, 100)[:, np.newaxis]
y = np.sin(X).ravel() + np.sin(6 * X).ravel() + rng.normal(0, 0.1, X.shape[0])
```

```python
import matplotlib.pyplot as plt

plt.plot(X, y)
plt.xlabel("X")
plt.ylabel("y")
```

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.tree import DecisionTreeRegressor

regr1 = DecisionTreeRegressor(max_depth=4)

regr2 = AdaBoostRegressor(DecisionTreeRegressor(max_depth=4), n_estimators=400, random_state=rng)

regr1.fit(X,y)
regr2.fit(X, y)

y1 = regr1.predict(X)
y2 = regr2.predict(X)
```

```python
import seaborn as sns

colors = sns.color_palette("colorblind")

plt.figure()
plt.scatter(X, y, color=colors[0], label = "training samples")
plt.plot(X, y1, color = colors[1], label = "n est = 1", linewidth = 2 )
plt.plot(X, y2, color = colors[2], label = "n est = 300", linewidth = 2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Ada boosted DTR")
plt.legend()
plt.show()
```
